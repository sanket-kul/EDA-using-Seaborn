### Separate Target from Features

```python
y = data.diagnosis                             # Target
drop_col = ['Unnamed: 32', 'id', 'diagnosis']
x = data.drop(drop_col, axis=1)                # Features

# We are dropping cols so axis=1, if we are dropping rows then axis=0
```

```python
x.head()
```

#### Check the data is Balanced or not

```python
y.value_counts(normalize=True)
```

As there is dissimiliar distribution of Benign and Malignant values, **Data is Not Balanced.**

### Diagnosis Distribution Visualization (Countplot)

```python
# Plot the distribution of target variables

ax = sns.countplot(x = 'diagnosis', data = data, label='Count')
B, M = y.value_counts()
print(f'Number of Benign tumor: {B}')
print(f'Number of Malignant tumor: {M}')
```

```python
x.describe()
```

### Visualizing Standardized Data with Seaborn (violinplot)

```python
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:, 0:10]], axis=1)

data = pd.melt(data, id_vars='diagnosis', var_name='features', value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features', y='value', hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45);
```

### Observing the Distribution of Values and their Variance with Swarm Plots (Swarmplot)


```python
warnings.filterwarnings('ignore')

sns.set(style='whitegrid', palette='muted')
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:, 0:10]], axis=1)

data = pd.melt(data, id_vars='diagnosis', var_name='features', value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features', y='value', hue='diagnosis', data=data)
plt.xticks(rotation=45);
```
